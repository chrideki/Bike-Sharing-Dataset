## Regression model using statsmodels.formula.api with two variables

```python
import statsmodels.formula.api as smf
linear_model = smf.ols(formula='cnt ~ weathersit',
                       data=bike)
fitted_model = linear_model.fit()
fitted_model.summary()
```

## Regression model using multiple variables

```python
import statsmodels.formula.api as smf
linear_model = smf.ols(formula='cnt ~ weathersit + temp + atemp + hum + windspeed',
                       data=bike)
fitted_model = linear_model.fit()
fitted_model.summary()
```

## Regression model using statsmodels.api

```python
import statsmodels.api as sm
X = pd.DataFrame({'cnt' : bike['weathersit']})
X = sm.add_constant(X)
y = bike['cnt']
## X['weight-squared'] = X['weight'].apply(lambda x: x*x) #explicitly makes new column
regressor = sm.OLS(y,X)
regressor = regressor.fit()
regressor.summary()
```

## Model regression with categorical variables

```python
import statsmodels.formula.api as smf
linear_model = smf.ols(formula='cnt ~ C(weathersit) + temp + atemp + hum + windspeed',
                       data=bike)
fitted_model = linear_model.fit()
fitted_model.summary()
```
